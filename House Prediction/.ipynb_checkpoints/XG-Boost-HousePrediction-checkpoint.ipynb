```python
#Step 1 - Loading Data
import pandas as pd
train_file_path = "input/train.csv"
house_data = pd.read_csv(train_file_path)
```

```python
house_data.describe()
```

```python
#Step 2- Filtering Columns
house_data.columns
```

```python
##input data
#house_predictors = ['LotArea','OverallQual','OverallCond','YearBuilt' ,'1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr', 'KitchenAbvGr',
#       'TotRmsAbvGrd', 'YrSold', 'GrLivArea']
#X = house_data[house_predictors]
X = house_data.drop(["SalePrice","Id"],axis=1)

X.head()
```

```python
## target data
Y = house_data.SalePrice
Y.head()
```

```python
#train_test_spilt
from sklearn.model_selection import train_test_split

train_X, val_X, train_Y, val_Y = train_test_split(X, Y,random_state = 0)
```

```python
#dropping columns with missing values
cols_with_missing = [col for col in train_X.columns 
                                 if train_X[col].isnull().sum() > 10]
print(train_X.columns)
print(cols_with_missing)
reduced_X_train = train_X.drop(cols_with_missing, axis=1)
reduced_X_val  = val_X.drop(cols_with_missing, axis=1)
print(reduced_X_train.shape)
print(reduced_X_val.shape)
```

```python
# filling in values with imputation

from sklearn.preprocessing import Imputer
def get_imputed_predictors(reduced_X_train,reduced_X_val):
    my_imputer = Imputer()
    imputed_X_train = my_imputer.fit_transform(reduced_X_train)
    imputed_X_val = my_imputer.transform(reduced_X_val)
    return imputed_X_train,imputed_X_val
```

```python
one_hot_encoded_training_predictors = pd.get_dummies(reduced_X_train)
one_hot_encoded_test_predictors = pd.get_dummies(reduced_X_val)
final_train, final_test = one_hot_encoded_training_predictors.align(one_hot_encoded_test_predictors,
                                                                    join='left', 
                                                                    axis=1)
```

```python
#imputation support

imputed_X_train,imputed_X_val  = get_imputed_predictors(final_train,final_test)
```

```python
#XG Boost

from xgboost import XGBRegressor

my_model = XGBRegressor(n_estimators=1000,learning_rate=0.05,objective='reg:linear')
# Add silent=True to avoid printing out updates with each cycle
my_model.fit(imputed_X_train, train_Y,verbose=False)
```

```python
# make predictions
predictions = my_model.predict(imputed_X_val)

from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, val_Y)))
```

```python
#feature importance weights
print(my_model.feature_importances_)
```

```python
#feature importance plot
from matplotlib import pyplot
pyplot.bar(range(len(my_model.feature_importances_)), my_model.feature_importances_)
pyplot.show()
```

```python
# Fit model using each importance as a threshold
from numpy import sort
import numpy as np
from sklearn.feature_selection import SelectFromModel
thresholds = sort(my_model.feature_importances_)
unique_thre = np.unique(thresholds)
for thresh in unique_thre:
    # select features using threshold
    selection = SelectFromModel(my_model, threshold=thresh, prefit=True)
    select_X_train = selection.transform(imputed_X_train)
    # train model
    selection_model = XGBRegressor(n_estimators=1000,learning_rate=0.05,objective='reg:linear')
    selection_model.fit(select_X_train, train_Y,verbose=False)
    # eval model
    select_X_test = selection.transform(imputed_X_val)
    y_pred = selection_model.predict(select_X_test)
    print("Thresh=%.3f, n=%d" % (thresh, select_X_train.shape[1]))
    print("Mean Absolute Error : " + str(mean_absolute_error(y_pred, val_Y)))
```
