```python
#Step 1 - Loading Data
import pandas as pd
train_file_path = "input/train.csv"
house_data = pd.read_csv(train_file_path)
```

```python
house_data.describe()
```

```python
#Step 2- Filtering Columns
house_data.columns
```

```python
## target data
Y = house_data.SalePrice
Y.head()
```

```python
##input data
house_predictors = ['LotArea','OverallQual','OverallCond','YearBuilt' ,'1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr', 'KitchenAbvGr',
       'TotRmsAbvGrd', 'YrSold', 'GrLivArea']
X = house_data[house_predictors]
X.head()
```

```python
##train validation data split

from sklearn.model_selection import train_test_split

train_X, val_X, train_Y, val_Y = train_test_split(X, Y,random_state = 0)
```

```python
#building model
from sklearn.tree import DecisionTreeRegressor

#Define model
house_model = DecisionTreeRegressor()

#Fit model
house_model.fit(train_X,train_Y)
```

```python
#Predictions and Accuracy

from sklearn.metrics import mean_absolute_error

predicted_home_prices = house_model.predict(val_X)
mean_absolute_error(val_Y,predicted_home_prices)
```
