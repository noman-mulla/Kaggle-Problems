```python
#Step 1 - Loading Data
import pandas as pd
train_file_path = "input/train.csv"
house_data = pd.read_csv(train_file_path)
```

```python
house_data.describe()
```

```python
#Step 2- Filtering Columns
house_data.columns
```

```python
## target data
Y = house_data.SalePrice
Y.head()
```

```python
##input data
house_predictors = ['LotArea','OverallQual','OverallCond','YearBuilt' ,'1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr', 'KitchenAbvGr',
       'TotRmsAbvGrd', 'YrSold', 'GrLivArea']
X = house_data[house_predictors]
X.head()
```

```python
##train validation data split

from sklearn.model_selection import train_test_split

train_X, val_X, train_Y, val_Y = train_test_split(X, Y,random_state = 0)
```

```python
#building model
from sklearn.tree import DecisionTreeRegressor

#Define model
house_model = DecisionTreeRegressor()

#Fit model
house_model.fit(train_X,train_Y)
```

```python
#Predictions and Accuracy

from sklearn.metrics import mean_absolute_error

predicted_home_prices = house_model.predict(val_X)
mean_absolute_error(val_Y,predicted_home_prices)
```

```python
def get_mae(max_leaf_nodes,train_X,train_Y,val_X,val_Y):
    my_model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state=0)
    my_model.fit(train_X,train_Y)
    my_model_prediction = my_model.predict(val_X)
    my_model_error = mean_absolute_error(val_Y,my_model_prediction)
    return my_model_error
```

```python
#comparing different tree depths

for max_leaf_nodes in [5,50,100,500,1000,2500,5000]:
    my_mae = get_mae(max_leaf_nodes,train_X,train_Y,val_X,val_Y)
    print("Max Leaf Nodes %d \t\t Mean absolute error %d" %(max_leaf_nodes,my_mae))

```

```python
# RandomForest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def get_randomForest_mae(max_depth,train_X,train_Y,val_X,val_Y):
    forest_model = RandomForestRegressor(max_depth = max_depth,random_state=0)
    forest_model.fit(train_X,train_Y)
    forest_pred = forest_model.predict(val_X)
    return mean_absolute_error(val_Y,forest_pred)
```

```python
for max_depth in [5,50,100,500]:
    my_random_mae = get_randomForest_mae(max_depth,train_X,train_Y,val_X,val_Y)
    print("Max Leaf Nodes %d \t\t Mean absolute error %d" %(max_depth,my_random_mae))
```

```python
##model submitted on kaggle into method to test accuracy via different column filling strategy
def getScore(train_X,train_Y,val_X,val_Y):
    forest_model = RandomForestRegressor()
    forest_model.fit(train_X,train_Y)
    forest_pred = forest_model.predict(val_X)
    return mean_absolute_error(val_Y,forest_pred)
    
```

```python
Y = house_data.SalePrice
X = house_data.drop(["SalePrice"],axis=1)
numeric_predictors = X.select_dtypes(exclude=['object'])

train_X, val_X, train_Y, val_Y = train_test_split(numeric_predictors, Y,random_state = 0)
```

```python
#dropping columns with missing values
cols_with_missing = [col for col in train_X.columns 
                                 if train_X[col].isnull().any()]
reduced_X_train = train_X.drop(cols_with_missing, axis=1)
reduced_X_val  = val_X.drop(cols_with_missing, axis=1)
print("Mean Absolute Error from dropping columns with Missing Values:")
print(getScore(reduced_X_train, train_Y,reduced_X_val,  val_Y))
```

```python
# filling in values with imputation

from sklearn.preprocessing import Imputer

my_imputer = Imputer()
imputed_X_train = my_imputer.fit_transform(train_X)
imputed_X_val = my_imputer.transform(val_X)
print("Mean Absolute Error from Imputation:")
print(getScore(imputed_X_train,train_Y ,imputed_X_val, val_Y))
```

```python
#with Showing what was imputed

imputed_X_train_plus = train_X.copy()
imputed_X_val_plus = val_X.copy()

cols_with_missing = (col for col in train_X.columns 
                                 if train_X[col].isnull().any())
for col in cols_with_missing:
    imputed_X_train_plus[col + '_was_missing'] = imputed_X_train_plus[col].isnull()
    imputed_X_val_plus[col + '_was_missing'] = imputed_X_val_plus[col].isnull()

# Imputation
my_imputer = Imputer()
imputed_X_train_plus = my_imputer.fit_transform(imputed_X_train_plus)
imputed_X_val_plus = my_imputer.transform(imputed_X_val_plus)

print("Mean Absolute Error from Imputation while Track What Was Imputed:")
print(getScore(imputed_X_train_plus, train_Y,imputed_X_val_plus, val_Y))
```
